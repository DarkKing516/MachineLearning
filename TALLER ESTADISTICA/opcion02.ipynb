```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# **Variables Cuantitativas Opción 01**

```python
# Variables Cuantitativas Opción 01
estudiantes = [
    22, 30, 33, 34, 35, 30, 29, 27, 30, 31,
    25, 30, 33, 32, 34, 35, 33, 34, 30, 22,
    28, 26, 27, 28, 29, 22, 23, 24, 25, 26,
    27, 29, 28, 27, 26, 24, 26, 30, 33, 34,
    26, 35, 34, 30, 32, 33, 30, 31, 27, 28
]

dataEstudiantes = pd.DataFrame(estudiantes, columns=["estudiantes"])
```

* **Tabla Frecuencias**

```python
frecuencia = dataEstudiantes['estudiantes'].value_counts().reset_index()
frecuencia
```

```python
frecuencia.columns=['Matriculas', 'Frecuencia Absoluta']

frecuencia['Frecuencia Acumulada Absoluta'] = frecuencia['Frecuencia Absoluta'].cumsum()

frecuencia
```

```python
suma_adsoulta = frecuencia['Frecuencia Absoluta'].sum()

frecuencia['Frecuencia Relativa'] = frecuencia['Frecuencia Absoluta'] / suma_adsoulta

frecuencia['Frecuencia Relativa Acumulada'] = frecuencia['Frecuencia Relativa'].cumsum()

frecuencia
```

* **Graficas**

```python
plt.figure(figsize=(10, 5))
plt.bar(frecuencia['Matriculas'], frecuencia['Frecuencia Absoluta'], color='blue', alpha=0.7)
plt.xlabel('Matriculas')
plt.ylabel('Frecuencia Absoluta')
plt.title('Frecuencia Absoluta')
plt.grid(axis='y')
plt.show()
```

```python
plt.hist(frecuencia['Frecuencia Acumulada Absoluta'], bins=4, edgecolor='black', color="green")

plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencia Acumulada')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(frecuencia['Matriculas'], frecuencia['Frecuencia Relativa'], color='green', alpha=0.7)
plt.xlabel('Matriculas')
plt.ylabel('Frecuencia Relativa')
plt.title('Frecuencia Relativa')
plt.grid(axis='y')
plt.show()
```

```python
# Graficar Frecuencia Relativa Acumulada
plt.figure(figsize=(10, 5))
plt.plot(frecuencia['Matriculas'], frecuencia['Frecuencia Relativa Acumulada'], marker='o', linestyle='-', color='purple')
plt.xlabel('Matriculas')
plt.ylabel('Frecuencia Relativa Acumulada')
plt.title('Frecuencia Relativa Acumulada')
plt.grid(axis='y')
plt.show()
```

* **media, mediana, mínimo, máximo y moda.**

```python
media = dataEstudiantes.mean()
mediana = dataEstudiantes.median()
minimo = dataEstudiantes.min()
maximo = dataEstudiantes.max()
moda = dataEstudiantes.mode().iloc[0]

print("Estadísticas de dataEstudiantes:")
print(f"Media:{media}")
print(f"Mediana:{mediana}")
print(f"Mínimo:{minimo}")
print(f"Máximo:{maximo}")
print(f"Moda:{moda}")
```
